#### Task 1: Write a program that asks the user about the number of values he/she wants to enter.Than enter the values as per the required number, calculate its sum and identify the smallest value among them.The sample output is as follow:

```python
def main():
    try:
        
        n = int(input("Enter the number of values you want to enter: "))
        
        if n <= 0:
            print("Please enter a positive integer for the number of values.")
            return
 
        values = []
        
       
        for i in range(n):
            value = float(input(f"Enter value {i + 1}: "))
            values.append(value)
        
        
        total_sum = sum(values)
        
        
        smallest_value = min(values)
        
      
        print(f"The sum of the entered values is: {total_sum}")
        print(f"The smallest value among them is: {smallest_value}")
    
    except ValueError:
        print("Invalid input. Please enter numeric values.")

if __name__ == "__main__":
    main()
```

#### Task 2: The factorial function is used frequently in probability problems.The factorial of a positive integer n (written n! and pronounced “n factorial”) is equal to the product of the positive integers from 1 to n. Write a function factorial that accepts an integer as parameter and returns its factorial. Using the factorial function, write a program that evaluates the factorials of the integers from 1 to 5. Print the results in tabular format as following

```python
def factorial(n):
    """Returns the factorial of the integer n."""
    if n == 0 or n == 1:
        return 1
    else:
        result = 1
        for i in range(2, n + 1):
            result *= i
        return result

def main():
   
    print(f"{'Number':<10}{'Factorial':<10}")
    print("-" * 20)

    for i in range(1, 6):
        print(f"{i:<10}{factorial(i):<10}")

if __name__ == "__main__":
    main()
```

```python
def main():

    secret_number = 4
    
    print("Welcome to the number-guessing game!")
    
    while True:
        try:
            
            guess = int(input("Enter your guess: "))

            if guess == secret_number:
                print("Congratulations! You guessed the secret number!")
                break
            else:
                print("Wrong guess. Try again!")
        
        except ValueError:
            print("Please enter a valid integer.")

if __name__ == "__main__":
    main()
```

#### Task 4: The greatest common divisor (GCD) of two integers is the largest integer that evenly divides each of the two numbers. Write function gcd that returns the greatest common divisor of two integers.

```python
def gcd(a, b):
    """Returns the greatest common divisor of two integers a and b."""
    while b != 0:
        a, b = b, a % b
    return a

def main():
    
    try:
        num1 = int(input("Enter the first integer: "))
        num2 = int(input("Enter the second integer: "))
 
        result = gcd(num1, num2)
        
        print(f"The GCD of {num1} and {num2} is: {result}")
    
    except ValueError:
        print("Please enter valid integers.")

if __name__ == "__main__":
    main()
```

#### Task5: An integer is said to be prime if it is divisible only by 1 and itself.For example, 2, 3, 5 and 7 are prime, but 4, 6, 8 and 9 are not. 
##### a) Write a function that determines if a number is prime. 
##### b) Use this function in a program that determines and prints all the prime numbers between 1 and 10,000.


##### 

```python
def is_prime(n):
    """Check if a number is a prime number."""
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True
def print_primes_up_to(max_num):
    """Print all prime numbers up to max_num."""
    for num in range(1, max_num + 1):
        if is_prime(num):
            print(num)

print_primes_up_to(10000)
```

#### Task6: Write a code that prints on screen all the 4-digit Armstrong numbers.

```python
def is_armstrong(n):
    s = str(n)
    return n == sum(int(digit) ** 4 for digit in s)

for num in range(1000, 10000):
    if is_armstrong(num):
        print(num)
```
